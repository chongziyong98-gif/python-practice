```python
#Day 5 æœ€ç»ˆæŒ‘æˆ˜é¢˜ï¼šè¿·ä½ ç‚¹é¤ç³»ç»Ÿï¼ˆæ•´åˆå›žé¡¾é¡¹ç›®ï¼‰

print("=== æ¬¢è¿Žä½¿ç”¨è¿·ä½ ç‚¹é¤ç³»ç»Ÿ ===")

menu = {
    "burger": 8.5,
    "fries": 3.5,
    "cola": 2.0,
    "icecream": 4.0,
    "salad": 5.0
} #ä¸ºä»€ä¹ˆè¦åˆ†è¡Œå†™ï¼Ÿå› ä¸ºè¿™æ ·æ›´æ–¹ä¾¿ä¿®æ”¹ è¿™å«åš implicit line joiningï¼ˆéšå¼æ¢è¡Œï¼‰ã€‚å•è¡Œä¹Ÿæ˜¯åˆæ³•çš„

cart = {}  # ç”¨å­—å…¸è®°å½•é¤åä¸Žæ•°é‡

while True: #while ç”¨åœ¨å½“ä¸ç¡®å®šç”¨æˆ·è¦è¾“å…¥å¤šå°‘æ¬¡æ—¶ï¼Œfor ç”¨åœ¨æœ‰é¢„è®¾å›ºå®šæ¬¡æ•°çš„å¾ªçŽ¯
    print("\n--- ä¸»èœå• ---")
    print("è¾“å…¥ menu æŸ¥çœ‹èœå•")
    print("è¾“å…¥ add ç‚¹é¤")
    print("è¾“å…¥ cart æŸ¥çœ‹è´­ç‰©è½¦")
    print("è¾“å…¥ checkout ç»“è´¦å¹¶å¯ä¿®æ”¹")
    print("è¾“å…¥ exit ç¦»å¼€ç³»ç»Ÿ")

    command = input("è¯·è¾“å…¥æŒ‡ä»¤: ")
    if command.lower().strip() == "menu": #.lower()å˜æˆå°å†™ï¼Œ.strip()åˆ é™¤ç©ºè¡Œ
      print(f"\n===ä¸»èœå•===")
      for food, price in menu.items(): #menu.items() = éåŽ†menu. for æ˜¯æ¡ä»¶å¾ªçŽ¯ã€‚äºŒ food,priceæ˜¯è‡ªå®šä¹‰ï¼Œå¯ä»¥æ›¿æ¢æˆå…¶ä»– (æ˜¯itemsä¸æ˜¯item)
        print(f"{food}:${price:.2f}")#:.2f åªèƒ½ç”¨åœ¨printé‡Œé¢ï¼Œå› ä¸ºå®ƒæ˜¯ç”¨æ¥ä¿®é¥°print stringçš„


    elif command.lower().strip() == "add":
      food = input("è¯·è¾“å…¥è¦ç‚¹çš„é¤åï¼š ")
      if food not in menu: # if xx not in list æ£€æŸ¥ä¸œè¥¿åœ¨ä¸åœ¨é¢„è®¾åˆ—è¡¨
        print("èœå•ä¸­æ²¡æœ‰è¯¥é¤å“ï¼Œè¯·é‡æ–°è¾“å…¥ã€‚")
        continue #è·³è¿‡è¿™æ¬¡ï¼Œå¹¶è®©é¡¾å®¢é‡æ–°è¾“å…¥ (è¿™é‡Œä¼šå›žåˆ°ä¸Šä¸€å±‚food input)
      try: #è¿™è¾¹å…ˆç”¨tryæ˜¯å› ä¸ºè¾“å…¥æ•°é‡æœ‰ä¸‰ç§ç»“æžœï¼Œè¾“å…¥>1 æ•°å­—ï¼Œè¾“å…¥<1æ•°å­—ï¼Œä»¥åŠè¾“å…¥å­—ä¸².ç”±äºŽif åœ¨ tryçš„ä¸‹å±‚ï¼Œæ‰€ä»¥å½“ç”¨æˆ·è¾“å…¥é”™è¯¯æ•°é‡æ—¶å¯ä»¥ç›´æŽ¥ä¿®æ”¹
        qty = int(input("è¯·è¾“å…¥æ•°é‡"))
        if qty < 1:
          print("æ•°é‡å¿…é¡»å¤§äºŽ0!")
          continue #è¿™é‡Œcontinurä¼šä»Žqtyå¼€å§‹
      except ValueError:
        print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡æ–°è¾“å…¥")
        continue #è¿™é‡Œcontinurä¼šä»Žqtyå¼€å§‹
      cart[food] = cart.get(food, 0) + qty #å–å‡ºcart ä¸­ food ç›®å‰çš„æ•°é‡ï¼ˆå¦‚æžœæ²¡è¿™ä¸ªé”®å°±å½“ä½œ0ï¼Œå¦‚æžœå°‘äº†0ï¼Œé‚£å®ƒä¼šè¿”å›žnoneï¼Œå°±ä¸èƒ½+ä¸Šäº†ï¼‰ï¼Œå†åŠ ä¸Šè¿™æ¬¡è¦åŠ å…¥çš„ qtyã€‚
      print(f"âœ… å·²å°† {qty} ä»½ {food} åŠ å…¥è´­ç‰©è½¦ï¼")

    elif command == "cart":
      if not cart: # if notå°±æ˜¯ æ£€æŸ¥æŽ¥ä¸‹æ¥çš„cartæ˜¯å¦æ²¡æœ‰ä»»ä½•èµ„æ–™
        print("ðŸ›’ è´­ç‰©è½¦ä¸ºç©º")
      else:
        print("\n--- å½“å‰è´­ç‰©è½¦ ---")
        total = 0 #ç”±äºŽæŽ¥ä¸‹æ¥è¦ç»™å‡ºåˆ—è¡¨ä¸­è®¢è¿‡çš„æ€»æ•°ï¼Œæ•…å…ˆé¢„è®¾ä¸º0
        for food, qty in cart.items(): #è¿™é‡Œå®šä¹‰food,qtyåœ¨cartï¼Œä¸€å®šè¦å’Œä¸Šé¢ç»Ÿä¸€(foodæ˜¯é‡å¤ä½¿ç”¨åœ¨cartå’Œmenuçš„)
          price = menu[food] * qty
          total += price
          print(f"{food} x{qty} = ${price:.2f}")
        print(f"æ€»ä»·: ${total:.2f}")

    elif command == "checkout":
        if not cart:
          print("è´­ç‰©è½¦ä¸ºç©ºï¼Œæ— æ³•ç»“è´¦ã€‚")
          continue
        print("\n=== ç»“è´¦æ¸…å• ===")
        total = 0
        for food, qty in cart.items():
            price = menu[food] * qty
            total += price
            print(f"{food} x{qty} = ${price:.2f}")
        print(f"æ€»ä»·: ${total:.2f}")

        choice = input("æ˜¯å¦è¦ä¿®æ”¹è®¢å•ï¼Ÿ(yes/no): ").lower().strip() #è¿™é‡Œç»™é¡¾å®¢é€‰æ‹©
        if choice == "yes":
            print("è¿”å›žèœå•ä¸­ï¼Œè¯·ä¿®æ”¹è®¢å•ã€‚")
            continue
        else: #å°ç‘•ç–µï¼Œè¿™é‡Œä»£è¡¨çš„æ˜¯yesä»¥å¤–çš„ä»»ä½•è¾“å…¥ï¼Œæ‰€ä»¥å°±ç®—ä¹±è¾“å…¥ä¹Ÿèƒ½ç»§ç»­ã€‚ç”¨whileå°±èƒ½è§£å†³
            print("ðŸ’° æ„Ÿè°¢å…‰ä¸´ï¼äº¤æ˜“å®Œæˆï¼")
            break
    elif command == "exit":
        print("ç³»ç»Ÿå·²é€€å‡ºã€‚")
        break

    else:
        print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡è¯•ã€‚") #è¿™é‡Œæ˜¯commandä¹±å†™åŽè¾“å‡ºçš„æ–‡å­—
```

```python
#è¿›é˜¶é¢˜ â‘ ï¼šåŠ¨æ€å¥½å‹åå•ï¼ˆå¯æ–°å¢žï¼ç§»é™¤ï¼æ˜¾ç¤ºï¼‰

friendlist = []

while True:
  command = input("è¯·è¾“å…¥æŒ‡ä»¤(æ·»åŠ å¥½å‹:add/ç§»é™¤å¥½å‹:remove/æ£€è§†åˆ—è¡¨:list/ç»“æŸæ‰§è¡Œ:end)")
  if command.lower() == "end":
    print("ç»“æŸæ‰§è¡Œ")
    break
  elif command.lower() == "add":
    friend = input("è¯·è¾“å…¥å¥½å‹å§“å: ")
    friendlist.append(friend)
    print(f"å½“å‰å¥½å‹åˆ—è¡¨:{friendlist}")
  elif command.lower() == "remove":
    remove_friend = input("è¯·è¾“å…¥è¦ç§»é™¤çš„å¥½å‹: ")
    if remove_friend in friendlist:
      friendlist.remove(remove_friend)
      print(f"å½“å‰å¥½å‹åˆ—è¡¨:{friendlist}")
    else:
      print("å¥½å‹ä¸å­˜åœ¨åˆ—è¡¨ï¼Œè¯·é‡æ–°è¾“å…¥")
  elif command.lower() == "list":
    print(f"å½“å‰å¥½å‹åˆ—è¡¨:{friendlist}")
  else:
    print("è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥")
```

```python
#ç»ƒä¹ é¢˜ 2ï¼šåº“å­˜ç®¡ç†ç³»ç»Ÿ

inventory = ["è‹¹æžœ", "é¦™è•‰", "æ©™å­"]

while True:
  item = input("è¯·è¾“å…¥æ“ä½œ (add/remove/list/end)ï¼š")
  if item.lower() == "add": #.lower() é»˜è®¤å°†è¾“å…¥å­—ç¬¦å˜å°å†™ï¼Œå¤§å°å†™åœ¨pythonä¼šå¯¼è‡´ä¸ä¸€æ ·ç»“æžœ
    add_item = input("è¯·è¾“å…¥è¦æ·»åŠ çš„å•†å“ï¼š")
    inventory.append(add_item) #.append åŠ ä¸Šä¸œè¥¿ç»§ç»­
    print(inventory) #è®°å¾—å¦‚æžœè¦ä¿ç•™appendï¼Œä¸è¦break
  elif item.lower() == "remove":
    remove_item = input("è¯·è¾“å…¥è¦ç§»é™¤çš„å•†å“ï¼š")
    if remove_item in inventory:   # âœ… æ”¹æˆ remove_item
        inventory.remove(remove_item)
        print(inventory)
    else:
        print("å•†å“ä¸å­˜åœ¨æ¸…å•ï¼Œè¯·é‡æ–°è¾“å…¥ï¼š")
  elif item.lower() == "list":
    print(inventory)
  elif item.lower() == "end":
    break
  else:
    print("è¾“å…¥é”™è¯¯ï¼Œè¯·é‡è¯•")
```

```python
#ç»ƒä¹ é¢˜ 1ï¼šè´­ç‰©ç»“ç®—å™¨

prices = []

for i in range(5):
  while True:
    price = input(f"è¯·è¾“å…¥ç¬¬{i+1}ä¸ªå•†å“ä»·æ ¼ï¼š")
    try:
      price = float(price)
      prices.append(price)
      break
    except ValueError :
      print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡æ–°è¾“å…¥!")

total = sum(prices)
average = total/len(prices)
above_average = [p for p in prices if p >= average] # pæ˜¯ä¸€ä¸ªä»£åï¼Œå¯æ›¿æ¢ã€‚è¿™å¥çš„æ„æ€æ˜¯(åœ¨priceså½“ä¸­ï¼Œå¦‚æžœp>=averageï¼Œé‚£pæœ‰ä»€ä¹ˆ)

print(f"æ‰€æœ‰å•†å“ä»·æ ¼ï¼š{prices}")
print(f"æ€»ä»·ï¼š{total}")
print(f"é«˜äºŽå¹³å‡ä»·çš„å•†å“ï¼š{above_average}")
```

```python
#é¢˜ç›®ä¸‰ï¼šå¹³å‡åˆ†è®¡ç®—å™¨ï¼ˆè¿›é˜¶ç‰ˆï¼‰


scores = []

for num in range(5):
  while True:
    score = input(f"è¯·è¾“å…¥ç¬¬{num+1}ä¸ªæ•°å­—: ")
    try:
      score = float(score)
      scores.append(score)
      break
    except ValueError:
      print("è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥")

average = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)

print(f"å¹³å‡åˆ†ä¸ºï¼š{average}")
print(f"æœ€é«˜åˆ†ä¸ºï¼š{highest_score}")
print(f"æœ€ä½Žåˆ†ä¸ºï¼š{lowest_score}")
```

```python
#æŒ‘æˆ˜é¢˜ 1ï¼šå¥‡å¶æ•°ç»Ÿè®¡å™¨

odd_numbers = []
even_numbers = []

for index in range(5):
  while True:
    user_input = input(f"è¯·è¾“å…¥ç¬¬{index+1}ä¸ªæ•°å­—:")
    try:
      user_input = int(user_input)
      if user_input%2 == 0:
        even_numbers.append(user_input)
        break
      else:
        odd_numbers.append(user_input)
        break
    except ValueError:
      print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡æ–°è¾“å…¥!")

count_odd = len(odd_numbers)
count_even = len(even_numbers)

print(f"å¶æ•°ä¸ªæ•°:{count_even}ï¼Œå¶æ•°åˆ—è¡¨ï¼š{even_numbers}")
print(f"å¥‡æ•°ä¸ªæ•°:{count_odd}ï¼Œå¥‡æ•°åˆ—è¡¨ï¼š{odd_numbers}")
```

```python
#Day5 æŒ‘æˆ˜é¢˜ #2ï¼šå­¦ç”Ÿæˆç»©ç»Ÿè®¡å™¨

scores = []

while True:
  score = input("è¯·è¾“å…¥å­¦ç”Ÿåˆ†æ•°ï¼ˆè¾“å…¥'end'ç»“æŸï¼‰:")
  if score.lower() == "end":
    break
  try:
    score = float(score)
    scores.append(score)
  except ValueError:
    print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡æ–°è¾“å…¥!")

average = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

pass_scores = [i for i in scores if i >= 60]
failed_scores = [i for i in scores if i < 60]

print(f"å¹³å‡åˆ†ï¼š{average}")
print(f"æœ€é«˜åˆ†ï¼š{highest}")
print(f"æœ€ä½Žåˆ†ï¼š{lowest}")
print(f"åŠæ ¼åˆ†æ•°ï¼š{pass_scores}")
print(f"ä¸åŠæ ¼åˆ†æ•°ï¼š{failed_scores}")
```
