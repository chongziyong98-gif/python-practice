```python
#练习题 3：库存统计器（进阶版）

inventory = ["苹果", "香蕉", "苹果", "橙子", "苹果"]

while True:
  check = input("请输入要查询的商品（输入'end'结束）：")
  if check.lower() == "end":#习惯优先把end程序放上面
    break
  if check in inventory:
    count = inventory.count(check) #.count(value) value不能是中文
    print(f"{check}一共有{count}个")
  else:
    print("商品不存在")
```

```python
#练习题 2：库存管理系统

inventory = ["苹果", "香蕉", "橙子"]

while True:
  item = input("请输入操作 (add/remove/list/end)：")
  if item.lower() == "add": #.lower() 默认将输入字符变小写，大小写在python会导致不一样结果
    add_item = input("请输入要添加的商品：")
    inventory.append(add_item) #.append 加上东西继续
    print(inventory) #记得如果要保留append，不要break
  elif item.lower() == "remove":
    remove_item = input("请输入要移除的商品：")
    if remove_item in inventory:   # ✅ 改成 remove_item
        inventory.remove(remove_item)
        print(inventory)
    else:
        print("商品不存在清单，请重新输入：")
  elif item.lower() == "list":
    print(inventory)
  elif item.lower() == "end":
    break
  else:
    print("输入错误，请重试")
```

```python
#练习题 1：购物结算器

prices = []

for i in range(5):
  while True:
    price = input(f"请输入第{i+1}个商品价格：")
    try:
      price = float(price)
      prices.append(price)
      break
    except ValueError :
      print("输入无效，请重新输入!")

total = sum(prices)
average = total/len(prices)
above_average = [p for p in prices if p >= average] # p是一个代名，可替换。这句的意思是(在prices当中，如果p>=average，那p有什么)

print(f"所有商品价格：{prices}")
print(f"总价：{total}")
print(f"高于平均价的商品：{above_average}")
```

```python
#题目三：平均分计算器（进阶版）


scores = []

for num in range(5):
  while True:
    score = input(f"请输入第{num+1}个数字: ")
    try:
      score = float(score)
      scores.append(score)
      break
    except ValueError:
      print("输入有误，请重新输入")

average = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)

print(f"平均分为：{average}")
print(f"最高分为：{highest_score}")
print(f"最低分为：{lowest_score}")
```

```python
#挑战题 1：奇偶数统计器

odd_numbers = []
even_numbers = []

for index in range(5):
  while True:
    user_input = input(f"请输入第{index+1}个数字:")
    try:
      user_input = int(user_input)
      if user_input%2 == 0:
        even_numbers.append(user_input)
        break
      else:
        odd_numbers.append(user_input)
        break
    except ValueError:
      print("输入无效，请重新输入!")

count_odd = len(odd_numbers)
count_even = len(even_numbers)

print(f"偶数个数:{count_even}，偶数列表：{even_numbers}")
print(f"奇数个数:{count_odd}，奇数列表：{odd_numbers}")
```

```python
#Day5 挑战题 #2：学生成绩统计器

scores = []

while True:
  score = input("请输入学生分数（输入'end'结束）:")
  if score.lower() == "end":
    break
  try:
    score = float(score)
    scores.append(score)
  except ValueError:
    print("输入无效，请重新输入!")

average = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

pass_scores = [i for i in scores if i >= 60]
failed_scores = [i for i in scores if i < 60]

print(f"平均分：{average}")
print(f"最高分：{highest}")
print(f"最低分：{lowest}")
print(f"及格分数：{pass_scores}")
print(f"不及格分数：{failed_scores}")
```
