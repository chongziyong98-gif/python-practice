```python
#é¢˜ç›®4:é¢˜ç›® 4ï¼šæˆç»©åˆ†æç³»ç»Ÿ (lists)

scores = [88, 56, 92, 74, 61, 100, 45, 83]

#è¾“å‡ºä»¥ä¸‹ç»“æœï¼š
#å¹³å‡åˆ†
#æœ€é«˜åˆ†ä¸æœ€ä½åˆ†
#åŠæ ¼äººæ•°ä¸ä¸åŠæ ¼äººæ•°
#å‰ä¸‰åæˆç»©ï¼ˆä»é«˜åˆ°ä½ï¼‰

average = sum(scores) / len(scores)
print(f"{average:.2f}")
maximum_score = max(scores)
print(maximum_score)
minimum_score = min(scores)
print(minimum_score)

passing_scores = [s for s in scores if s >= 60]
print(passing_scores)

top_three = sorted(scores,reverse=True)[:3]
print(top_three)


#é¢˜ç›® 5ï¼šå­¦ç”Ÿä¿¡æ¯è¡¨ï¼ˆDictionaryï¼‰

student = {"name": "Alice", "age": 20, "major": "CS"}

#æ–°å¢é”® "score"ï¼Œå€¼ä¸º 95
student["score"] = 95

#ä¿®æ”¹ "major" ä¸º "Data Science"
student["major"] = "Data Science"

#åˆ é™¤ "age"
del student["age"]

#ä½¿ç”¨ for å¾ªç¯æ‰“å°æ‰€æœ‰ key å’Œ value
for key, value in student.items():
  print(f"{key}:{value}")

#é¢˜ç›® 6ï¼šé›†åˆè¿ç®—ï¼ˆSupplierï¼‰
A = {"paper", "box", "tape", "glue"}
B = {"tape", "glue", "label", "carton"}

#ä¸¤å®¶å…±æœ‰å•†å“
print(A & B)
#ä»… A ä¾›åº”å•†å“
print(A - B)
#æ‰€æœ‰ä¸é‡å¤çš„å•†å“
print(A ^ B)
```

```python
#æŒ‘æˆ˜é¢˜ï¼šåº“å­˜ç®¡ç† & æŠ¥è¡¨ç³»ç»Ÿ (day1 - day9)

stock = {"apple": 50, "banana": 30, "orange": 20}

while True:
  command = input("è¯·è¾“å…¥æŒ‡ä»¤(æŸ¥è¯¢åº“å­˜:check / æ–°å¢äº§å“:add / æ›´æ–°æŸäº§å“æ•°é‡:update / æ˜¾ç¤ºæ‰€æœ‰åº“å­˜:show / ç®€æŠ¥:report / ç¦»å¼€:exit)").lower().strip()
  if command == "check":
    enquiry = input("è¯·è¾“å…¥è¦æŸ¥è¯¢çš„è´§å“åç§°: ").strip().lower()
    print(f"{enquiry}çš„æ•°é‡ä¸º: {stock.get(enquiry,"æ‰¾ä¸åˆ°è¯¥å•†å“")}")

  elif command == "add":
    item = input("è¯·è¾“å…¥è¦åŠ å…¥çš„å•†å“åç§°: ").lower().strip()
    if item in stock:
      print(f"å•†å“å·²å­˜åœ¨ï¼Œå½“å‰æ•°é‡ä¸º:{stock[item]}")
    else:
      quantity = int(input("è¯·è¾“å…¥æ·»åŠ æ•°é‡"))
      try:
        if quantity >= 0:
          stock[item] = quantity
          print(f"å·²æ·»åŠ {item},æ•°é‡ä¸º{quantity}")
        else:
          print("æ•°é‡ä¸èƒ½ä¸ºè´Ÿæ•°ï¼")
      except ValueError:
        print("è¾“å…¥æ— æ•ˆï¼Œè¯·è¾“å…¥æ•°å­—ï¼")

  elif command == "update":
    item = input("è¯·è¾“å…¥è¦æ›´æ–°çš„å•†å“åç§°: ").lower().strip()
    if item in stock:
      quantity = input("è¯·è¾“å…¥æ·»åŠ æ•°é‡(ç”¨-ä»£è¡¨å‡å°‘æ•°é‡):")
      try:
        change = int(quantity)
        stock[item] += change
        if stock[item] < 0:
          stock[item] = 0
          print("åº“å­˜ä¸èƒ½ä¸ºè´Ÿï¼Œå·²è®¾ä¸º0")
      except ValueError:
        print("è¾“å…¥æ— æ•ˆï¼Œè¯·è¾“å…¥æ•°å­—ï¼")

  elif command == "show":
    for item, quantity in stock.items():
      print(f"{item}:{quantity}")

  elif command == "report":
    item_count = len(stock)
    total_quantity = sum(stock.values())
    max_item = max(stock, key=stock.get) #æ‰¾å‡ºâ€œå€¼æœ€å¤§â€çš„ keyï¼ˆå³åº“å­˜æœ€å¤šçš„å•†å“ï¼‰
    max_quantity = stock[max_item]
    print("\nğŸ“Š åº“å­˜ç®€æŠ¥")
    print(f"ğŸ“¦ å½“å‰å…±æœ‰ {item_count} ç§å•†å“")
    print(f"ğŸ§® åº“å­˜æ€»é‡: {total_quantity}")
    print(f"ğŸ† åº“å­˜æœ€å¤šçš„æ˜¯ {max_item} ({max_quantity})\n")

  elif command == "exit":
    break

  else:
    print("è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥ï¼")
```
