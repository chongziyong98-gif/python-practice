```python
#day10: ç»ˆææŒ‘æˆ˜ï¼šå¤šå­¦ç”Ÿæˆç»©è¯„ä¼°ç³»ç»Ÿ

#ç”¨æˆ·è¾“å…¥å¤šä½å­¦ç”Ÿæˆç»©ï¼ˆæ ¼å¼å¦‚ä¸‹ï¼‰ å¼ ä¸‰:80,90,100; æå››:70,60,65; ç‹äº”:88,92,81
#è®¡ç®—æ¯ä½å­¦ç”Ÿçš„å¹³å‡åˆ†
#åˆ¤æ–­ç­‰çº§ï¼ˆç”¨ä½ ä¸Šä¸€é¢˜çš„ grade_system() å‡½æ•°ï¼‰
#æœ€åæ‰“å°æŠ¥å‘Šæ¸…å•
#æœ€é«˜åˆ†å­¦ç”Ÿ
#å¹³å‡ç­‰çº§ç»Ÿè®¡

def grade_system(score):
  if score >= 90:
    grade = "A"
  elif score >= 80:
    grade = "B"
  elif score >= 70:
    grade = "C"
  elif score >= 60:
    grade = "D"
  else:
    grade = "F"
  return grade

student_grade = {}
student_average_score = {}

while True:
  command = input("æ–°å¢å­¦ç”Ÿ(add) / ç»“æŸå¹¶å¼€å§‹è¿ç®—(exit): ").lower().strip()
  if command == "add":
    student_name = input(f"è¯·è¾“å…¥å­¦ç”Ÿ{len(student_list)+1}å§“å: ")
    user_input = input("è¯·è¾“å…¥ä»–çš„æˆç»©(ç”¨é€—å·åŒºåˆ†): ")
    try:
      scores_list = [float(i.strip()) for i in user_input.split(",")]
      average_score = round(sum(scores_list)/len(scores_list),2)
      grade = grade_system(average_score)
      print(f"{student_name}çš„å¹³å‡åˆ†ä¸º{average_score},ä»–çš„è¯„çº§ä¸º{grade}")
      student_grade[student_name] = grade
      student_average_score[student_name] = average_score
    except ValueError:
      print("è¾“å…¥æ— æ•ˆï¼Œè¯·è¾“å…¥æ•°å­—ï¼")

  elif command == "exit":
    if len(student_grade) == 0:
      print("æ•°æ®ä¸ºç©ºï¼Œè¯·å…ˆè¾“å…¥æ•°æ®!")
    else:
      break
  else:
    print("è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥ï¼")

# è¾“å‡ºæˆç»©æŠ¥å‘Š
print("\nğŸ“Š å­¦ç”Ÿæˆç»©æŠ¥å‘Š")
print("----------------------")
for name in student_grade:
  print(f"{name} å¹³å‡åˆ†: {student_average_score[name]} ç­‰çº§: {student_grade[name]}")

# ğŸ† æœ€é«˜åˆ†å­¦ç”Ÿ
top_student = max(student_average_score, key=student_average_score.get) #max(list,key=dict.get)ä»student_average_scoreé‡Œå–å‡ºkey(ç¬¬ä¸€ä¸ª)
print(f"\nğŸ† æœ€é«˜åˆ†å­¦ç”Ÿ: {top_student} ({student_average_score[top_student]}åˆ†)")

# ğŸ“ˆ å¹³å‡ç­‰çº§ç»Ÿè®¡
grade_values = list(student_grade.values()) #åˆ—å‡ºstudent_gradeé‡Œé¢çš„values(ç¬¬äºŒä¸ª)
total_students = len(grade_values)
grade_count = {g: grade_values.count(g) for g in set(grade_values)} #æŠŠ dict.values() å˜æˆ listï¼Œç„¶åå¯¹æ¯ä¸ªä¸åŒçš„ç­‰çº§ç”¨ .count() è®¡æ•°

print("\nğŸ“ˆ ç­‰çº§ç»Ÿè®¡æŠ¥å‘Š:")
for g, count in sorted(grade_count.items(), reverse=True):
  percent = round((count / total_students) * 100, 1)
  print(f"ç­‰çº§ {g}: {count}äºº ({percent}%)")

```
