```python
#Day 9 å­¦ä¹ ä¸»é¢˜ï¼šé›†åˆ Set ä¸å»é‡æ“ä½œ

fruits = {"apple", "banana", "orange", "apple"} #setçš„ç‰¹ç‚¹æ˜¯ç”¨{}ï¼Œä½†æ˜¯å†™æ³•å’Œlistç›¸åŒ
print(fruits)      # å¯¼å‡ºç»“æœä¼šè‡ªåŠ¨å¿½è§†é‡å¤å†…å®¹

A = {"apple", "banana", "cherry"}
B = {"banana", "cherry", "durian"}

print(A | B)   # å¹¶é›† union â†’ {'apple', 'banana', 'cherry', 'durian'} #ä¸¤ä¸ªsetåˆå¹¶ï¼Œç„¶åä¸ä¼šå‡ºç°é‡å¤
print(A & B)   # äº¤é›† intersection â†’ {'banana', 'cherry'} #åªæ˜¾ç¤ºé‡å é¡¹
print(A - B)   # å·®é›† difference â†’ {'apple'} #ç”¨a-bå‰©ä¸‹çš„ç»„åˆï¼ŒåŒç†ï¼Œb-aä¼šå‡ºç°durian
print(A ^ B)   # å¯¹ç§°å·®é›† symmetric_difference â†’ {'apple', 'durian'} #è¿™ä¸ªç¬¦å·æ˜¯ ^ keyboard shift+6

#ä»¥ä¸‹æ˜¯ç®€åŒ–setè¾“å…¥æ³•
student_a = input("è¯·è¾“å…¥Açš„å…´è¶£ï¼ˆç”¨é€—å·åˆ†éš”ï¼‰: ").lower().split(",") #æ”¶é›†èµ„æ–™ï¼Œè¿™é‡Œéœ€è¦userä¸»åŠ¨è¾“å…¥ï¼ŒåŒºåˆ†å…´è¶£. split(",") æ£€æµ‹ï¼Œåˆ†éš”
student_b = input("è¯·è¾“å…¥Bçš„å…´è¶£ï¼ˆç”¨é€—å·åˆ†éš”ï¼‰: ").lower().split(",")

set_a = set(i.strip() for i in student_a) #set()è‡ªåŠ¨æ•´åˆæˆset, i for i in student_a = è¯»å–student_aæ‰€æœ‰é¡¹ç›®ï¼Œå¥—ç”¨
set_b = set(i.strip() for i in student_b)
print("æ‰€æœ‰ç‹¬ç‰¹å…´è¶£ï¼š", set_a | set_b)
print("å…±åŒå…´è¶£ï¼š", set_a & set_b)
print("Aç‹¬æœ‰å…´è¶£ï¼š", set_a - set_b)
print("Bç‹¬æœ‰å…´è¶£ï¼š", set_b - set_a)
```

```python
#Day 9 ç»ƒä¹  1ï¼šå…´è¶£é‡å åˆ†æå™¨

#å»ºç«‹ä¸€ä¸ªç³»ç»Ÿï¼Œåˆ†æä¸¤ä½åŒå­¦çš„å…´è¶£ï¼Œè¾“å‡ºï¼š
#1ï¸âƒ£ ä¸¤äººéƒ½å–œæ¬¢çš„å…´è¶£ï¼ˆäº¤é›†ï¼‰
#2ï¸âƒ£ åªæœ‰Aå–œæ¬¢çš„å…´è¶£
#3ï¸âƒ£ åªæœ‰Bå–œæ¬¢çš„å…´è¶£
#4ï¸âƒ£ ä¸¤äººæ‰€æœ‰çš„å…´è¶£ï¼ˆå¹¶é›†ï¼‰

student_1 = input("è¯·è¾“å…¥å­¦ç”Ÿ1çš„å…´è¶£: ").lower().split(",")
student_2 = input("è¯·è¾“å…¥å­¦ç”Ÿ2çš„å…´è¶£: ").lower().split(",")

set_1 = set(i.strip() for i in student_1)
set_2 = set(i.strip() for i in student_2)

print(f"ä»–ä»¬éƒ½å–œæ¬¢:{set_1 & set_2}")
print(f"åªæœ‰å­¦ç”Ÿ1å–œæ¬¢{set_1 - set_2}")
print(f"åªæœ‰å­¦ç”Ÿ2å–œæ¬¢{set_2 - set_1}")
print(f"ä¸¤äººæ‰€æœ‰å…´è¶£:{set_1 | set_2}")
```

```python
#Day 9 è¿›é˜¶æŒ‘æˆ˜é¢˜ï¼šä¾›åº”å•†å¯¹æ¯”åˆ†æç³»ç»Ÿ

#ä½ ç°åœ¨æœ‰ä¸¤å®¶ä¾›åº”å•†ï¼ˆSupplier A ä¸ Supplier Bï¼‰ï¼Œ
#ä½ éœ€è¦ç³»ç»Ÿåœ°åˆ†æä»¥ä¸‹é¡¹ç›®ï¼š

#1ï¸âƒ£ ä¸¤å®¶å…±åŒä¾›åº”çš„äº§å“
#2ï¸âƒ£ ä»…Aä¾›åº”çš„äº§å“
#3ï¸âƒ£ ä»…Bä¾›åº”çš„äº§å“
#4ï¸âƒ£ ä¸¤å®¶åŠ èµ·æ¥çš„æ‰€æœ‰äº§å“ï¼ˆå»é‡ï¼‰
#5ï¸âƒ£ è®¡ç®—æ¯å®¶ç‹¬æœ‰äº§å“æ•°é‡ï¼Œè¾“å‡ºç®€å•åˆ†ææŠ¥å‘Š

supplier_list = []

while len(supplier_list) < 2: #ç”±äºç›®å‰è¿›åº¦åªæ”¯æŒ2ç»„å¯¹æ¯”ï¼Œæ•…æ­¤è¿›è¡Œwhileé™åˆ¶
  name = input(f"è¯·è¾“å…¥ç¬¬ {len(supplier_list) + 1} å®¶ä¾›åº”å•†åå­—: ").lower().strip()
  if name:
    supplier_list.append(name)
    print(f"âœ… å·²æ·»åŠ ä¾›åº”å•†: {name}")
  else:
    print("â— åå­—ä¸èƒ½ä¸ºç©ºï¼Œè¯·é‡æ–°è¾“å…¥ã€‚")

supplier_products = {} #setè‡ªå®šä¹‰è¦ç”¨{}

for supplier in supplier_list:
  supply_items = input(f"è¯·è¾“å…¥{supplier}æä¾›çš„å•†å“(ç”¨é€—å·åˆ†éš”): ").lower().split(",")
  supplier_products[supplier] = set(i.strip() for i in supply_items)


supplier_a, supplier_b = supplier_list #ä¸ºsupplier_listå®šä¹‰ç³»ç»Ÿå†…ä»£ç 
set_a = supplier_products[supplier_a]
set_b = supplier_products[supplier_b]

# é›†åˆåˆ†æ
common = set_a & set_b
only_a = set_a - set_b
only_b = set_b - set_a
all_items = set_a | set_b

# è¾“å‡ºæŠ¥å‘Š
print("\nğŸ“¦ ä¾›åº”å•†åˆ†ææŠ¥å‘Š")
print(f"å…±åŒä¾›åº”çš„äº§å“: {common}")
print(f"{supplier_a} ç‹¬å®¶ä¾›åº”: {only_a}")
print(f"{supplier_b} ç‹¬å®¶ä¾›åº”: {only_b}")
print(f"æ‰€æœ‰äº§å“(ä¸é‡å¤): {all_items}")

# ç»Ÿè®¡
print("\nğŸ“Š æ•°æ®ç»Ÿè®¡:")
print(f"{supplier_a} ç‹¬å®¶æ•°é‡: {len(only_a)}")
print(f"{supplier_b} ç‹¬å®¶æ•°é‡: {len(only_b)}")
print(f"å…±åŒä¾›åº”æ•°é‡: {len(common)}")
```
