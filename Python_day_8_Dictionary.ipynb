```python
#Day 8 ä¸»é¢˜ï¼šå­—å…¸ Dictionary

student = {
    "name": "John",
    "age": 19,
    "major": "Computer Science"
}

# 1ï¸âƒ£ æ–°å¢ä¸€é¡¹ "score" å€¼ä¸º 85
# 2ï¸âƒ£ ä¿®æ”¹å¹´é¾„ä¸º20
# 3ï¸âƒ£ åˆ é™¤ "major"
# 4ï¸âƒ£ è¾“å‡ºå®Œæ•´çš„ student å­—å…¸
# 5ï¸âƒ£ ä½¿ç”¨ for å¾ªç¯é€è¡Œæ˜¾ç¤ºæ‰€æœ‰å†…å®¹

student["score"] = 85 #ä¸éœ€è¦ç”¨åˆ°append
student["age"] = 20
del student["major"]
print(student)

for key, value in student.items(): #ä¸èƒ½å•å•åªå†™studentï¼Œ.items()æŠŠå­—å…¸å†…å®¹ã€Œæ‰“åŒ…æˆä¸€å¯¹å¯¹(key, value)ã€è®©ä½ ä¸€æ¬¡å–å‡º
  print(f"{key}: {value}")
```

```python
#ç»ƒä¹ 2

scores = {
    "alice": 85,
    "bob": 92,
    "charlie": 77,
    "david": 59
}

#è¾“å…¥å­¦ç”Ÿåå­—æ¥æŸ¥è¯¢åˆ†æ•°
#å¦‚æœæ‰¾ä¸åˆ°å­¦ç”Ÿï¼Œæç¤ºâ€œæ‰¾ä¸åˆ°è¯¥å­¦ç”Ÿâ€
#è¾“å…¥ add æ—¶å¯ä»¥æ–°å¢å­¦ç”Ÿä¸åˆ†æ•°
#è¾“å…¥ show æ˜¾ç¤ºå…¨éƒ¨å­¦ç”Ÿæˆç»©
#è¾“å…¥ exit ç¦»å¼€ç³»ç»Ÿ

while True:
  command = input("\nè¾“å…¥æŒ‡ä»¤(è¾“å…¥checkä»¥æŸ¥çœ‹æˆç»© / è¾“å…¥addæ–°å¢å­¦ç”Ÿæˆç»© / è¾“å…¥showæ˜¾ç¤ºå…¨éƒ¨å­¦ç”Ÿæˆç»© / è¾“å…¥exitç¦»å¼€ç³»ç»Ÿ): ")
  if command.lower().strip() == "add":
    name = input("è¯·è¾“å…¥å­¦ç”Ÿå§“å: ").lower().strip()
    score = float(input("è¯·è¾“å…¥ä»–çš„åˆ†æ•°")) #è®°å¾—ä½¿ç”¨float
    scores[name] = score
    print(f"å·²æ·»åŠ å­¦ç”Ÿ{name},åˆ†æ•°ä¸º{score}")

  elif command.lower().strip() == "check": #è¿™é‡Œä¸éœ€è¦ç”¨forå‡½æ•°åšå¾ªä¾‹ï¼Œå› ä¸ºæˆ‘ä»¬åªæ˜¯æ‰¾èµ„æ–™
    name = input("è¯·è¾“å…¥æƒ³å’¨è¯¢çš„å­¦ç”Ÿå§“å: ").lower().strip()
    print(scores.get(name,"æ‰¾ä¸åˆ°è¯¥å­¦ç”Ÿ"))

  elif command.lower().strip() == "show":
    print("\nå…¨éƒ¨å­¦ç”Ÿæˆç»©å¦‚ä¸‹:")
    for name,score in scores.items():
      print(f"{name}:{score}")

  elif command.lower().strip() == "exit":
    print("ç³»ç»Ÿç»“æŸè¿è¡Œ:")
    break

  else:
    print("è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥ï¼")
```

```python
#ç»ƒä¹ 2 è¯å…¸æŸ¥è¯¢ç³»ç»Ÿ

dictionary = {"apple": "è‹¹æœ", "banana": "é¦™è•‰", "cat": "çŒ«"}

#è¾“å…¥è‹±æ–‡å•è¯ â†’ è¾“å‡ºå¯¹åº”ä¸­æ–‡
#å¦‚æœæ‰¾ä¸åˆ° â†’ æç¤º "è¯å…¸ä¸­æš‚æ— è¯¥å•è¯"
#è¾“å…¥ "add" â†’ è®©ç”¨æˆ·æ·»åŠ æ–°å•è¯å’Œç¿»è¯‘
#è¾“å…¥ "exit" â†’ é€€å‡ºç³»ç»Ÿ

while True:
  command = input("è¯·è¾“å…¥æ“ä½œ(find æŸ¥æ‰¾å•è¯ / add æ–°å¢å•è¯ / exit é€€å‡ºç³»ç»Ÿ)").strip().lower()
  if command == "find":
    word = input("è¯·è¾“å…¥æƒ³æŸ¥æ‰¾çš„å•è¯: ")
    print(dictionary.get(word,"è¯å…¸ä¸­æš‚æ— è¯¥å•è¯"))

  elif command == "add":
    word = input("è¯·è¾“å…¥æ–°çš„å•è¯").lower().strip()
    if word in dictionary:
      print(f"å•è¯"{word}"å·²å­˜åœ¨ï¼Œå½“å‰æ„æ€æ˜¯ï¼š{dictionary[word]}")
    else:
      meaning = input("è¯·è¾“å…¥ç¿»è¯‘ç»“æœ")
      dictionary[word] = meaning
      print(f"æ–°å¢æˆåŠŸï¼Œ{word}çš„æ„æ€æ˜¯{meaning}")

  elif command == "exit":
    print("ç³»ç»Ÿç»“æŸè¿è¡Œ")
    break

  else:
    print("è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥ï¼")
```

```python
# day 8 è¿›é˜¶æŒ‘æˆ˜é¢˜: å•†å“åº“å­˜ç®¡ç†ç³»ç»Ÿï¼ˆDictionary + æ•°å€¼è¿ç®—ï¼‰

stock_list = {
    "pyrans_6mm": 50,
    "pyranstar_5mm": 30,
    "pyrans_8mm": 40,
    "pyrobelite_25mm": 60,
    "pyrobelite_52mm": 42
}

while True:
    command = input("\nè¯·è¾“å…¥æŒ‡ä»¤(æŸ¥è¯¢åº“å­˜: enquiry / æ–°å¢å•†å“: add / æ›´æ–°åº“å­˜æ•°é‡: update / æ˜¾ç¤ºæ‰€æœ‰åº“å­˜: show / é€€å‡ºç³»ç»Ÿ: exit): ").lower().strip()

    if command == "enquiry":
        item = input("è¯·è¾“å…¥è¦æŸ¥è¯¢çš„è´§å“åç§°: ").lower().strip()
        print(stock_list.get(item, "âŒ å•†å“ä¸å­˜åœ¨"))

    elif command == "add":
        add_item = input("è¯·è¾“å…¥è¦æ·»åŠ çš„å•†å“åç§°: ").lower().strip()
        if add_item in stock_list:
            print(f"âš ï¸ å•†å“å·²å­˜åœ¨ï¼Œåº“å­˜ä¸º {stock_list[add_item]}")
        else:
            try:
                item_quantity = int(input("è¯·è¾“å…¥äº§å“æ•°é‡: "))
                if item_quantity >= 0:
                    stock_list[add_item] = item_quantity
                    print(f"âœ… å·²æ·»åŠ  {add_item} ï¼Œæ•°é‡ä¸º {item_quantity}")
                else:
                    print("â— æ•°é‡ä¸èƒ½ä¸ºè´Ÿæ•°")
            except ValueError:
                print("âŒ è¾“å…¥æ— æ•ˆï¼Œè¯·è¾“å…¥æ•°å­—ï¼")

    elif command == "update":
        update_item = input("è¯·è¾“å…¥è¦æ›´æ–°çš„è´§å“åç§°: ").lower().strip()
        if update_item in stock_list:
            update_quantity = input("è¯·è¾“å…¥æ›´æ–°æ•°é‡(ä¾‹å¦‚ +10 æˆ– -5): ")
            try:
                change = int(update_quantity)
                stock_list[update_item] += change
                if stock_list[update_item] < 0:
                    stock_list[update_item] = 0
                    print("âš ï¸ åº“å­˜ä¸èƒ½ä¸ºè´Ÿï¼Œå·²è®¾ä¸º 0ã€‚")
                print(f"âœ… å·²æ›´æ–° {update_item} ï¼Œå½“å‰åº“å­˜ï¼š{stock_list[update_item]}")
            except ValueError:
                print("âŒ è¾“å…¥æ— æ•ˆï¼Œè¯·è¾“å…¥æ•°å­—æˆ–åŠ å‡å·ã€‚")
        else:
            print("âŒ å•†å“ä¸å­˜åœ¨ï¼Œè¯·å…ˆä½¿ç”¨ add æ·»åŠ ã€‚")

    elif command == "show":
        print("\nğŸ“¦ å½“å‰åº“å­˜åˆ—è¡¨ï¼š")
        for item, quantity in stock_list.items():
            print(f"{item}: {quantity}")

    elif command == "exit":
        print("ç³»ç»Ÿç»“æŸï¼Œæ„Ÿè°¢ä½¿ç”¨ ğŸ‰")
        break

    else:
        print("âŒ è¾“å…¥æœ‰è¯¯ï¼Œè¯·é‡æ–°è¾“å…¥ï¼")

```
