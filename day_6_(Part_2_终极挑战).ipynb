```python
# day_6 ç»ˆææŒ‘æˆ˜é¢˜ï¼šå­¦ç”Ÿæˆç»©åˆ†æç³»ç»Ÿï¼ˆè°ƒè¯•gradeï¼‰

#ç¬¬ä¸€æ­¥:å¯è®¾å®šçš„gradesç­‰çº§

def set_grade_threshold(): #è®©definitionæ”¾åœ¨æœ€å‰é¢ï¼Œåç»­formulaå¯ä»¥ç›´æ¥å¥—ç”¨
  print("ğŸ“ æ¬¢è¿ä½¿ç”¨å­¦ç”Ÿæˆç»©åˆ†æç³»ç»Ÿï¼")
  print("ä½ å¯ä»¥è‡ªå®šä¹‰æ¯ä¸ªç­‰çº§çš„æœ€ä½åˆ†æ•°ï¼ˆèŒƒå›´ 0-100ï¼‰")
  print("ä¾‹å¦‚ï¼šA=90, B=80, C=70, D=60, F=0")
  print("è‹¥æƒ³ä½¿ç”¨ç³»ç»Ÿé»˜è®¤å€¼ï¼Œè¯·ç›´æ¥è¾“å…¥ skipã€‚\n")

  default_thresholds = {
        "A": 90,
        "B": 80,
        "C": 70,
        "D": 60,
        "F": 0
  }

  thresholds = {} #è¿™é‡ŒåŒæ ·ä½¿ç”¨dictionary{}çš„æ‹¬å¼§ï¼Œæ”¾ç©ºæ˜¯"å¾…è¾“å…¥"

  for grade in ["A", "B", "C", "D", "F"]: #è¿™é‡Œè¡¨ç¤ºè¿™äº”ä¸ªæ˜¯gradeï¼Œæˆ‘ä»¬è¦æ±‚useråˆ†åˆ«è¾“å…¥è¿™äº”ç»„gradeçš„score)
    while True: #ç”±äºå¯èƒ½å‡ºç°è¾“å…¥é”™è¯¯ï¼Œæ•…æ­¤ç”¨whileä»¥é¿å…è¾“å…¥é”™è¯¯å¯¼è‡´ä¸­æ–­
      user_input = input(f"è¯·è¾“å…¥ç­‰çº§{grade}çš„æœ€ä½åˆ†æ•°(æˆ–è¾“å…¥skipä½¿ç”¨é»˜è®¤{default_thresholds[grade]}):") #default_threshold[grade]ï¼Œä»default_thresholdä¸­ï¼Œæ‰¾å‡º[grade](ä¹Ÿå°±æ˜¯æœ¬æ¬¡å¾ªç¯)æ‰€"ç¿»è¯‘"çš„æ„æ€

      if user_input.lower().strip() == "skip":
        thresholds[grade] = default_thresholds[grade] #å…ˆå‰å®šä¹‰äº†gradeä¸ºABCDF(forå¾ªç¯çš„é‚£ä¸€è¡Œ),è¿™é‡Œçš„æ„æ€æ˜¯è®¾å®šthreshold dictionaryé‡Œçš„gradeä¸ºdefault_threshold dictionaryé‡Œçš„grade
        break #è®°å¾—break

      try: #è¿™é‡Œçš„ä¸Šä¸€è¡Œä¸éœ€è¦å†ç”¨elseäº†ï¼Œç›´æ¥try exceptå°±è¡Œ
        value = float(user_input)
        if 0 <= value <= 100:
          thresholds[grade] = value #è¿™é‡Œå®šä¹‰thresholdé‡Œé¢gradeæ‰€ä»£è¡¨çš„æ„æ€ï¼Œè®°å¾—ï¼Œthreshold[grade]æ˜¯ä½¿ç”¨[]ï¼Œè€Œä¸æ˜¯{}
          break #è®°å¾—åœ¨æˆåŠŸè®¾å®šååŠ ä¸Šbreak
        else:
          print("è¯·è¾“å…¥0-100ä¹‹é—´çš„åˆ†æ•°!")
      except ValueError:
        print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡æ–°è¾“å…¥!")

  print("\næˆåŠŸè®¾å®šç­‰çº§é—¨æ§›å¦‚ä¸‹:") #è¿™ä¸ªä¼šåœ¨æ‰€æœ‰5ä¸ªgradeè¾“å…¥æˆåŠŸåæ˜¾ç¤º(è®°å¾—åœ¨foråŒè¡Œ)
  for grade, score in thresholds.items():
    print(f"{grade}:{score}åˆ†ä»¥ä¸Š")
  print("\n") #æ¢è¡Œ
  return thresholds

# ==== ç¬¬äºŒæ­¥ï¼šè®¡ç®—æˆç»©ç­‰çº§ ====

def get_grade(score, thresholds): #thresholds æ˜¯ä»ç¬¬ä¸€æ­¥è¿”å›æ¥çš„dictionary
  for grade, min_score in thresholds.items(): #.items()ä¼šæŠŠå­—å…¸è½¬æˆå¯è¿­ä»£çš„é”®å€¼å¯¹ï¼ˆkey-value pairsï¼‰, è¡¨ç¤ºï¼šä¾åºå–å‡ºè¿™äº› key å’Œ value
    if score >= min_score: #å½“åˆ†æ•° â‰¥ å½“å‰ç­‰çº§çš„é—¨æ§›
      return grade
  return "F" #å…œåº•æ¡ä»¶


# ==== ç¬¬ä¸‰æ­¥ï¼šä¸»ç¨‹åº ====

while True:
  thresholds = set_grade_threshold() #è¿™é‡Œé‡æ–°å®šä¹‰thresholds, ç›®å‰æ˜¯è®©åˆ†æ•°åŒºåˆ†æŒ‰ç…§å·²ç»è°ƒè¯•å¥½äº†çš„gradeç³»ç»Ÿèµ°
  student_score_list = [] #é¢„è®¾ä»¥æ–¹ä¾¿æ¥ä¸‹æ¥appendè¢«è¾“å…¥çš„åˆ†æ•°ï¼Œå¹¶ç”¨äºè®¡ç®—average,highest and lowest
  student_grade_list = [] #åŒä¸Šï¼Œä¸è¿‡è¿™æ¬¡æ˜¯appendä¸ªåˆ«ç­‰çº§

  while True:
    student_count = len(student_score_list) #å®šä¹‰æ–°çš„student_countä»¥è®©æ¥ä¸‹æ¥userçŸ¥é“ä»–æ­£åœ¨è¾“å…¥ç¬¬å‡ ä¸ªå­¦ç”Ÿ
    student_score = input(f"è¯·è¾“å…¥ç¬¬{student_count + 1}ä¸ªå­¦ç”Ÿçš„åˆ†æ•° (è¾“å…¥exitç»“æŸ): ")

    if student_score.lower().strip() == "exit":
      break

    try:
      student_score = float(student_score)
      if 0 <= student_score <= 100:
        g = get_grade(student_score,thresholds) #å®šä¹‰gæ˜¯æ–¹ä¾¿æ¥ä¸‹æ¥è¾“å…¥ï¼Œget_grade(student_score,thresholds)æ˜¯ç¬¬äºŒæ­¥å®šä¹‰çš„å‡½æ•°
        print(f"åˆ†æ•°{student_score}çš„ç­‰çº§æ˜¯{g}")
        student_score_list.append(student_score)#è¿™é‡Œæ²¡æœ‰é—®é¢˜
        student_grade_list.append(g)#è¿™é‡Œä¹Ÿæ˜¯gæ²¡æœ‰é—®é¢˜ï¼Œå› ä¸ºgåœ¨ä¸Šé¢è¢«å®šæˆgradeäº†
      else:
        print("è¾“å…¥æ— æ•ˆï¼Œè¯·è¾“å…¥0-100ä¹‹é—´çš„æ•°å­—!")
    except ValueError:
      print("è¾“å…¥æ— æ•ˆï¼Œè¯·é‡æ–°è¾“å…¥!")

# ==== ç¬¬å››æ­¥ï¼šç»Ÿè®¡åˆ†æ ====
  if len(student_score_list) == 0: #è¿™é‡Œæ”¾åœ¨ç¬¬äºŒè¡Œå¼€å§‹ï¼Œå› ä¸ºæˆ‘ä»¬æ¥ä¸‹æ¥è¦ç»™choice userè®©ä»–ä»¬å†³å®šè¦ä¸è¦ç»§ç»­
    print("æ²¡æœ‰è¾“å…¥ä»»ä½•æˆç»©ã€‚") #è¡¥å……ä¸€ç‚¹ï¼Œå¦‚æœè¿™ä¸€æ­¥åœ¨ç¬¬ä¸€è¡Œï¼Œé‚£ä¹ˆå½“useræ²¡æœ‰è¾“å…¥ä»»ä½•ä¸œè¥¿ï¼Œç¨‹åºå°†å¡æ­»
  else:
    total_students = len(student_grade_list)
    average_score = sum(student_score_list) / len(student_score_list)
    highest_score = max(student_score_list)
    lowest_score = min(student_score_list)

    print("\n======== æˆç»©ç»Ÿè®¡æŠ¥å‘Š ========")
    print(f"ğŸ‘¥ æ€»äººæ•°: {total_students}")
    print(f"ğŸ“Š å¹³å‡åˆ†: {average_score:.2f}")
    print(f"ğŸ† æœ€é«˜åˆ†: {highest_score}")
    print(f"ğŸ“‰ æœ€ä½åˆ†: {lowest_score}")

    print("\nğŸ“˜ ç­‰çº§åˆ†å¸ƒ:")
    for g in ["A", "B", "C", "D", "F"]:
      print(f"{g}: {student_grade_list.count(g)} äºº")

    # å‰ä¸‰å
    print("\nğŸ¥‡ å‰ä¸‰åæˆç»©:")
    top_scores = sorted(student_score_list, reverse=True)[:3] #sorted() ä¼šæŠŠä¸€ä¸ªåˆ—è¡¨æ’åºï¼Œäº§ç”Ÿæ–°çš„åˆ—è¡¨ï¼ˆé»˜è®¤æ˜¯å°-å¤§ã€‚reverse=Trueè¡¨ç¤ºdecendingã€‚[:3]åˆ‡ç‰‡å–3ä¸ª
    for i, score in enumerate(top_scores, start=1): #enumerate()è®©ä½ åœ¨ for å¾ªç¯ä¸­åŒæ—¶æ‹¿åˆ°ç´¢å¼•ï¼ˆåºå·ï¼‰ä¸å€¼,start=1 è®©ä½ ä»1å¼€å§‹ï¼Œå› ä¸ºé»˜è®¤æ˜¯0
      print(f"{i}. {score}")

# ==== ç¬¬äº”æ­¥ï¼šæ˜¯å¦é‡å¯ ====
  choice = input("\næ˜¯å¦è¦é‡æ–°å¼€å§‹ç³»ç»Ÿï¼Ÿ(yes/no): ").lower().strip()
  if choice != "yes":
    print("ç³»ç»Ÿç»“æŸï¼Œæ„Ÿè°¢ä½¿ç”¨ ğŸ‰")
    break

```
