```python
# === 练习任务 Day 3: BMI 计算器 ===

print("=== 练习任务 Day 3: BMI 计算器 ===")

weight = float(input("请输入体重（kg）："))
height = float(input("请输入身高（m）："))

# 检查输入是否有效 (把无效的代码放前面)
if height <= 0 or weight <= 0:
    print("输入无效，请输入正数！")
else:
    BMI = weight / (height ** 2) # 计算 BMI

    print(f"你的 BMI 为 {BMI:.2f}")# 输出结果，:.2f 表示保留2位小数

    if BMI < 18.5:
        print("这表示你的体重过轻，要注意饮食！")
    elif 18.5 <= BMI <= 24.9:
        print("这表示你的体重正常，保持健康！")
    else:
        print("这表示你的体重过重，要注意饮食！")
```

```python
print("=== 练习任务 Day 3: 分数等级判定系统 ===")

score = float(input("请输入分数: "))

if score<0 or score>100:
  print("输入无效")
else:
  if score >= 90:
    print("Grade A")
  elif 80 <= score <= 89: #这边用 score >= 80 就好，同下
    print("Grade B")
  elif 70 <= score <= 79:
    print("Grade C")
  elif 60 <= score <= 69:
    print("Grade D")
  else:
    print("Failed")
```

```python
print("=== 练习任务 Day 3: 简单购物折扣计算器 ===")

amount=float(input("请输入购买数额: "))


if amount <= 100:
  discount_rate = 0
  final_price = amount * (1 - discount_rate)
  print(f"原价为:${amount}\n折扣:{discount_rate*100}%\n应付金额:${final_price}") #discount_rate 可以*100然后接上%以让结果更美观
elif amount <= 499:
  discount_rate = 0.05
  final_price = amount * (1 - discount_rate)
  print(f"原价为:${amount}\n折扣:{discount_rate*100}%\n应付金额:${final_price}")
elif amount <= 999:
  discount_rate = 0.1
  final_price = amount * (1 - discount_rate)
  print(f"原价为:${amount}\n折扣:{discount_rate*100}%\n应付金额:${final_price}")
else:
  discount_rate = 0.2
  final_price = amount * (1 - discount_rate)
  print(f"原价为:${amount}\n折扣:{discount_rate*100}%\n应付金额:${final_price}")
```
