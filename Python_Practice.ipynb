```python
#Variable = Value
name='Alvin Chong' #字符串string必须用引号括起来，可以使用单引号''或双引号"
age=27
Age=28 #remember capital letter and small letter can be different
hobby='diving'

print('hello,world') # comma(,)separates outputs; print adds space automatically
print('my name is',name) # once a variable is defined, you don't need quotes when printing it
print('I am',age,'years old for this year'', '"and will be",Age,"next year") #to input comma, use ',' instead.
print('I like',hobby)


#错误写法： print("my name is",name. "I am",age,"years old this year, and will be",Age,'next year.')
#1: .句号不是连接点
#错误写法2： print("My name is", name,"I am", age, "years old this year, and will be", Age, "next year.") = My name is Alvin chong I am 27 years old this year, and will be 28 next year.
#2: 在print()里面，每个逗号（，）会自动有空格

print("\n----- Method 1: Using plus sign ( + ) -----")
print("My name is",name+".","I am",age,"years old this year, and will be",Age,"next year.")#使用 + 达成没有空格的衔接

print("\n----- Method 2: Using f-string -----")
print(f"My name is {name}. I am {age} years old this year, and will be {Age} next year.")
#f-string 进阶用法 用{}在盒子

print("\n----- To key multiple layer in one line -----")
print(f"Hello!\nMy name is {name}. I am {age} this year, and will be {Age} next year.\nI like {hobby}.")
# \n 是“换行符号”，所以method 1 不管用
```
