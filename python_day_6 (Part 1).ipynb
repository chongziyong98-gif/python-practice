```python
#练习 4（进阶挑战）：成绩等级统计系统


def grade(scores):
  if scores >= 90: #记得不要写<100，因为这样的话就会永远是A
    return "A" #直接表达要return什么就好了
  elif scores >= 80:
    return "B"
  elif scores >= 70:
    return "C"
  elif scores >= 60:
    return "D"
  else:
    return "F"

student_grades = [] #先预设

while True:
  student_count = len(student_grades) + 1 #如果没有+1，他会从0开始
  score = input(f"请输入学生{student_count}成绩(或输入exit结束): ")
  if score.lower().strip() == "exit":
    break
  else:
    try:
      score = float(score)
      if 0 <= score <=100: #正常分数时
        score = grade(score)
        student_grades.append(score)

      else: #可能会输入负数或 >100 的数值
        print("请输入 0-100 之间的分数")

    except ValueError: #避免输入数字以外的数字然后报错
      print("输入无效，请输入数字")
      continue

# 统计每个等级人数
print("\n=== 成绩等级分布 ===")

for g in ["A", "B", "C", "D", "F"]: #直接定义g (或者grade或是别的)，这样一来下面不用一个个输入
  print(f"{g}: {student_grades.count(g)} 人") #list要计算单个grade有多少个时，用list.count(g) 上一行已经定义g了

print("\n统计完毕，谢谢使用！")


```

```python
#Day 6 进阶练习②：函数版 BMI 计算器

def BMI_Calculator(weight,height): #这里def BMI_Calculator()，weight,height 需要和下面的算式内容同步
  BMI = weight/(height**2)
  if BMI < 18.5: #算好BMI了，还没有return之前还能继续加其他需要return的东西
    status = "体重过轻"
  elif BMI < 25:
    status = "体重正常"
  else:
    status = "体重过重"
  return BMI,status #可以return好几个数值，只要前面写好

while True:
  wei = input("请输入体重(KG)或输入exit结束: ") #也可以写成w或任何，只要下面使用函数的内容相等于这里定好的数值就好
  if wei.lower().strip() == "exit":
    break
  else:
    try:
      wei = float(wei)
    except ValueError:
      print("输入有误，请重新输入!")
      continue
  hgh = input("请输入身高(M): ") #建议写一个看得懂的，方便后续调整
  if hgh.lower().strip() == "exit":
    break
  else:
    try:
      hgh = float(hgh)
    except ValueError:
      print("输入有误，请重新输入!")
      continue

  BMI,statu = BMI_Calculator(wei,hgh) #这里也是定义。我故意将status写错以证明没有问题。重点在于def好的顺序(1,2)。同理，甚至可以写成statu,BMI
  print(f"你的BMI为{BMI:.2f},属于{statu}。\n") #这里的输出必须和上一行定义好的同步哦
```

```python

#Day 6 强化挑战题：BMI 平均统计器


def BMI_Calculator(weight,height): #这里def BMI_Calculator()，weight,height 需要和下面的算式内容同步
  BMI = weight/(height**2)
  if BMI < 18.5: #算好BMI了，还没有return之前还能继续加其他需要return的东西
    status = "体重过轻"
  elif BMI < 25:
    status = "体重正常"
  else:
    status = "体重过重"
  return BMI,status #可以return好几个数值，只要前面写好

BMI_records = []


while True:
  wei = input("请输入体重(KG)或输入exit结束: ") #也可以写成w或任何，只要下面使用函数的内容相等于这里定好的数值就好
  if wei.lower().strip() == "exit":
    break
  else:
    try:
      wei = float(wei)
    except ValueError:
      print("输入有误，请重新输入!")
      continue
  hgh = input("请输入身高(M): ") #建议写一个看得懂的，方便后续调整
  if hgh.lower().strip() == "exit":
    break
  else:
    try:
      hgh = float(hgh)
    except ValueError:
      print("输入有误，请重新输入!")
      continue
  BMI,statu = BMI_Calculator(wei,hgh)
  print(f"你的BMI为{BMI:.2f},属于{statu}。\n") #这里的输出必须和上一行定义好的同步哦
  BMI_records.append(BMI) #这里记得在括弧里放想要加入的数值

if len(BMI_records) == 0:
  print("没有任何有效的BMI记录，系统结束。")
else:
  calculate_times = len(BMI_records)
  average_BMI = sum(BMI_records) / len(BMI_records)
  highest_BMI = max(BMI_records)
  lowest_BMI = min(BMI_records)

  print("======== 统计报告 ========")
  print(f"共计算 {calculate_times} 次")
  print(f"平均BMI: {average_BMI:.2f}")
  print(f"最高BMI: {highest_BMI:.2f}")
  print(f"最低BMI: {lowest_BMI:.2f}")
```
