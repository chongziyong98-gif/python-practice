```python
#练习任务 Day 4: 循环版 BMI 计算器

while True: #while 条件(可以是count或是其他自定义条件)
  height = input("请输入身高M: ")
  if height.lower() == "exit": #加上.lower()，即使用户输入Exit, 也能自动转换成全小写
    print("测试结束")
    break
  if not height.replace('.', '', 1).isdigit(): #.replace('.', '', 1) replace=替换什么.,替换的数值，替换几次。.isdigit=检查剩下的是不是数字
    print("输入无效，请重新输入")
    continue #这个continue可以在user写错时可以让系统跳转回去，否则会报错
  weight = input("请输入体重KG: ")
  if not weight.replace("." ,"", 1).isdigit(): #.replace('.', '', 1) replace=替换什么.,替换的数值，替换几次。.isdigit=检查剩下的是不是数字
    print("输入无效，请重新输入")
    continue


  height = float(height)
  weight = float(weight)
  BMI = weight / (height**2)

  print(f"你的BMI是{BMI:.2f}")
  if BMI < 18.5:
      print("体重过轻")
  elif BMI < 25:
      print("体重正常")
  else:
      print("体重过重")
print() #可以换行，让下一个user更好分辨
```

```python
#练习1：重复问候机

count=1 #while格式里的先决条件

while count <= 3:
  print("Hello, Alvin!")
  count += 1 #这个很重要，没有它就会无限循环，电脑无限问候你


#练习2：1 加到 N 的总和计算器

N = int(input("请输入一个整数"))
total = 0

for index in range(1,N+1): #range(start,stop,step),N+1作为stop 是因为stop不包含stop本身
  total += index
print(total)

#练习3：倒数计时器

num = int(input("请输入倒计时:"))
total = 0

for index in range(num,0,-1):
    print(f"倒计时{index}")
print("倒计时结束")
```

```python
#Day 4 进阶题 ①：平均分计算器

count = 0
total = 0

while True:
  score = input("请输入学生分数或输入exit结算")
  if score.lower() == "exit": #先从break开始
    break
  if not score.replace(".","",1).isdigit():#验证可以置放在同一行
    print("输入无效,请重新输入")
    continue #记得验证要有continue，意味这回到while开始循环
  total += float(score)
  count += 1

if count >= 1: #算式应该要在loop之外进行，除了那些需要从loop即时获得的数值
  average = total/count
  print(f"平均分为{average}")
else:
  print("没有输入任何分数") #防止除以0时的报错
```
