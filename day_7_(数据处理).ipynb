```python
#day 7: 主题：列表进阶 & 数据处理入门

#练习1:

student_score_list = [56, 78, 45, 89, 90, 32, 67, 100]

total = sum(student_score_list)
average = sum(student_score_list)/len(student_score_list)
highest_score = max(student_score_list)
lowest_score = min(student_score_list)


print(f"总分:{total}")
print(f"平均分:{average}")
print(f"最高分:{highest_score}")
print(f"最低分:{lowest_score}")

passing_scores = []
A_scores = []

for s in student_score_list: #这里以s(score) 代表list里面的内容
  if s >= 60:
    passing_scores.append(s)

for s in student_score_list: #这里以s(score) 代表list里面的内容，不会和上面冲突，因为上面for loop已经完成
  if s >= 90:
    A_scores.append(s)

print(f"及格分数为:{passing_scores}")
print(f"优异分数为:{A_scores}")

failed = [s for s in student_score_list if s < 60] #这个是更简短的说法，用[]直接list failed
print(failed)
```

```python
#Day 7 小挑战 — 成绩数据分析器

scores = [90, 82, 77, 59, 93, 68, 55, 100]

average_score = sum(scores)/len(scores)
highest = max(scores)
lowest = min(scores)
passed = [s for s in scores if s >= 60]
failed = [s for s in scores if s < 60]

scores_above_average = [s for s in scores if s >= average_score]
top_three_student = sorted(scores,reverse=True)[:3] #这里使用,reverse=True就能倒数了

print(f"平均分为{average_score}")
print(f"最高分为{highest},最低分为{lowest}")
print(f"及格分数为{passed}")
print(f"不及格分数为{failed}")
print(f"平均分以上的分数为{scores_above_average}")
print(f"前三名分数为{top_three_student}")
```
